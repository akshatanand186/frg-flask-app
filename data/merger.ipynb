```python
import pandas as pd

# Read in the data
img = pd.read_csv('images.csv')

#rename filename to id
img.rename(columns={'filename':'id'}, inplace=True)

#convert id to int64
img['id'] = img['id'].str.replace('.jpg', '').astype('int64')

img.head()
```

```python
# leave out inconsistent rows
style = pd.read_csv('styles.csv', on_bad_lines='skip')

style['id'] = style['id'].astype('int64')

style.head()
```

```python
# join both the tables on filename and id
df = pd.merge(img, style, on='id')
df.head()
```

```python
# save to csv
df.to_csv('merged.csv', index=False)
```
